```python
from google.colab import drive
import csv
import os
from translate import Translator

ROOT = '/content/drive' # location mount drive at

def mount_drive(ROOT):
    drive.mount(ROOT, force_remount=True)

mount_drive(ROOT)
```

```python
lang=['Turkish','Chinese','French','German']
typeArr=['train.csv','val.csv']
dbg=True

for i in range(len(lang)):
    translator= Translator(to_lang=lang[i])
    if dbg==True:
        translation = translator.translate("one")
        print(translation)
    for t in typeArr:
        with open(f'/content/drive/MyDrive/2020Paper_Reduced_Code/data/csv/English/{t}', mode="r") as csv_file:
            reader = csv.reader(csv_file) 
            try:
                os.mkdir(f'/content/drive/MyDrive/2020Paper_Reduced_Code/data/csv/{lang[i]}/')
            except Exception as e:
                print(e)
            with open(f"/content/drive/MyDrive/2020Paper_Reduced_Code/data/csv/{lang[i]}/og_{t}", mode="w") as new_file:
                writer_obj = csv.writer(new_file, delimiter=",") # Writes to the new CSV file 
                for id,data in enumerate(reader):
                    if not id==0:
                        data[2] = translator.translate(data[2])
                    writer_obj.writerow(data)
                new_file.close()
            csv_file.close()
```

```python
lang=['tr','zh-cn','de'] # french is giving error 
folder_name=['Turkish','Chinese','German']
'''
    tr=Turkish
    zh-cn=Chinese(simplified)
'''
typeArr=['train.csv','val.csv']
dbg=True

for i in range(len(lang)):
    translator= Translator()
    if dbg==True:
        translation = translator.translate("one",dest=f'{lang[i]}')
        print(translation)
    for t in typeArr:
        with open(f'/content/drive/MyDrive/2020Paper_Reduced_Code/data/csv/English/{t}', mode="r") as csv_file:
            reader = csv.reader(csv_file) 
            try:
                os.mkdir(f'/content/drive/MyDrive/2020Paper_Reduced_Code/data/csv/{folder_name[i]}/')
            except Exception as e:
                print(e)
            with open(f"/content/drive/MyDrive/2020Paper_Reduced_Code/data/csv/{folder_name[i]}/google_{t}", mode="w") as new_file:
                writer_obj = csv.writer(new_file, delimiter=",") # Writes to the new CSV file 
                for id,data in enumerate(reader):
                    if not id==0:
                        result=translator.translate(data[2],dest=f'{lang[i]}')
                        data[2] = result.text
                    writer_obj.writerow(data)
                new_file.close()
            csv_file.close()
```
